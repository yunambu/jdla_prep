## OpenCV

```python
import cv2 # OpenCV をインポート
```

```python
img = cv2.imread('data/sample.png') # 最初から Numpy 型で読み込み
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img) # OpenCV の場合、Blue -> Green -> Red の順でデフォルト画像を見る
```

```python
# BGR -> RGB の順に変更
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # cvt = convert
```

```python
plt.imshow(img_rgb)
```

## グレースケール変換

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_gray)
```

```python
plt.gray() # グレースケールで表示することをまず宣言
plt.imshow(img_gray)
```

```python
img_gray.shape
```

## フィルターをかける

```python
# kernel = フィルターのこと
kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```python
img_conv = cv2.filter2D(img_gray, 0, kernel) # 0 = 方向の指定で 0 の場合横向き
```

```python
plt.gray()
plt.imshow(img_conv)
```

```python
kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])
```

```python
img_conv = cv2.filter2D(img_gray, 0, kernel)
```

```python
plt.gray()
plt.imshow(img_conv)
```

```python
kernel = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
])
```

```python
img_conv = cv2.filter2D(img_gray, 0, kernel)
plt.gray()
plt.imshow(img_conv)
```
